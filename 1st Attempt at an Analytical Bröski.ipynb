```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 4
L = 1
N = 1000
dx = n*np.pi*L/N

x = [0]
ps = [0]
ps1 = [1]
ps2 = [0] 

for i in range(0,N):
    p = ps[-1] + dx * ps1[-1]
    p1 = ps1[-1] + dx * ps2[-1]
    p2 = ps2[-1] - dx * ps1[-1]
    px = x[-1] + dx/(n*np.pi)
    
    ps.append(p)
    ps1.append(p1)
    ps2.append(p2)
    x.append(px)

xa = np.array(x)
psi = np.array(ps)
psi1 = np.array(ps1)
psi2 = np.array(ps2)
analytical = np.sin(n*xa)

plt.figure(0)
plt.plot(x,psi,'r',label=psi)
plt.plot(x,psi1,'g',label=psi1,alpha=0.5)
plt.plot(x,psi2,'b',label=psi2,alpha=0.5)
plt.legend(["Wavefunction","1st Derviative","Second Derivative"],loc=4)

Error = np.abs(analytical - psi)

plt.figure(1)
plt.plot(x,Error)
plt.title("Error")
```
