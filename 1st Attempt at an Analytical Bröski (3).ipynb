```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1                                                                    #Principal Quantum Number    
L = 1                                                                    #Length of [THE BOX]
N = 10000                                                               #Number of points
norm = np.sqrt(2/L)
dx = n*np.pi/N                                                         #Small change in x

x = [0]                                                                  #Initial conditions
ps = [0]                                                                 ##
ps1 = [norm]                                                                ##
ps2 = [0]                                                                ##

for i in range(0,N):                                                     #[LÖÖP]
    p2 = ps2[-1] - ps1[-1]*dx
    p1 = ps1[-1] + p2*dx
    p =  ps[-1]  + p1*dx

    px = x[-1] + L/N                                                   #Small Change in x
    
    ps.append(p)                                                         #Me 'n the boys appending our lists
    ps1.append(p1)                                                       ##
    ps2.append(p2)                                                       ##
    x.append(px)                                                         ##

xa = np.array(x)                                                         #Converting lists to Numpy arrays
psi = np.array(ps)                                                       ##
psi1 = np.array(ps1)                                                     ##
psi1 = psi1*n*np.pi/L
psi2 = np.array(ps2)                                                     ##
psi2 = psi2*((n*np.pi/L)**2)
analytical = norm*np.sin(n*np.pi*xa/L)                                          #Analytical Soln

plt.figure(0)                                                            #Plotting the numerical wavefunction
plt.plot(x,psi,'r',label=psi)                                            #and its 1st and 2nd derivatives
plt.plot(x,psi1,'g',label=psi1,alpha=0.5)                                ##
plt.plot(x,psi2,'b',label=psi2,alpha=0.5)                                ##
plt.legend(["Wavefunction","1st Derviative","Second Derivative"],loc=4)  ##

Error = np.abs(analytical - psi)                                         #Creating an error array
                                                                         ##
plt.figure(1)                                                            #And plotting it
plt.plot(x,Error)                                                        ##
plt.title("Error")                                                       ##

plt.figure(2)
Energy = np.sum(psi2)/np.sum(psi)
```
